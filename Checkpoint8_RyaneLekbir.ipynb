```python
import requests as re
from bs4 import BeautifulSoup as bs
```

```python
link = "https://en.wikipedia.org/wiki/PlayStation_5"
req_link = re.get(link)
soup = bs(req_link.content, "html.parser")
print(soup.title)
print(soup.prettify())
```

```python
def link(link):
    details = soup.find("table",{"class":"infobox"})
    for i in details:
        row = i.find_all("tr")
        for k in row:
            heading = k.find_all("th")
            detail = k.find_all("td")
            for x,y in zip(heading, detail):
                print("{} :: {}".format(x.text, y.text))
                print("-----------------------------")

link(link)
```

```python
def claimLink(url):
    req1 = re.get(url)
    bs1 = bs(req1.content, "html.parser")
    
    findLink = bs1.find_all("a")
    link_List = []
    
    for i in findLink:
        link = i.get("href")
        link_List.append(link)
    
    return link_List

claimLink("https://en.wikipedia.org/wiki/PlayStation_5")
```
